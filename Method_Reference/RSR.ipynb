```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import norm

def rsr_analysis(data, weight=None, threshold=0.3, ascending=None):
    """
    RSR (Rank Sum Ratio) 分析法完整实现
    
    参数:
    ----------
    data : 2D array或DataFrame
        原始数据矩阵 (n_samples, n_indicators)
    weight : array, optional
        指标权重 (None表示等权重)
    threshold : float
        分档阈值 (默认0.3)
    ascending : array, optional
        各指标方向性 (True表示越大越好，False反之)
        
    返回:
    ----------
    result : DataFrame
        包含秩次、RSR值、分档结果的表格
    """
    # 数据预处理
    X = np.array(data, dtype=float)
    n, m = X.shape
    if ascending is None:
        ascending = np.ones(m, dtype=bool)  # 默认所有指标越大越好
    if weight is None:
        weight = np.ones(m) / m  # 等权重
    
    # 指标归一化 (方向性处理)
    X_norm = np.zeros_like(X)
    for i in range(m):
        if ascending[i]:
            X_norm[:, i] = (X[:, i] - X[:, i].min()) / (X[:, i].max() - X[:, i].min() + 1e-10)
        else:
            X_norm[:, i] = (X[:, i].max() - X[:, i]) / (X[:, i].max() - X[:, i].min() + 1e-10)
    
    # 计算秩次 (平均秩法处理相同值)
    R = np.zeros_like(X)
    for i in range(m):
        ranked = pd.Series(X_norm[:, i]).rank(method='average')
        R[:, i] = ranked.values
    
    # 计算RSR值
    RSR = (R.sum(axis=1) / (n * m)).reshape(-1, 1)
    
    # 概率单位转换
    probit = norm.ppf(RSR / (n + 1)) + 5  # 避免除零
    
    # 分档处理
    bins = _auto_bins(RSR, threshold)
    grades = np.digitize(RSR.flatten(), bins)
    grade_labels = [f'Grade {i+1}' for i in range(len(bins)-1)]
    
    # 结果整理
    result = pd.DataFrame({
        '样本': [f'Sample {i+1}' for i in range(n)],
        **{f'指标{j+1}_秩次': R[:, j] for j in range(m)},
        'RSR值': RSR.flatten(),
        'Probit值': probit.flatten(),
        '分档': pd.Categorical.from_codes(grades-1, categories=grade_labels)
    })
    
    # 可视化
    _plot_rsr(result, bins)
    
    return result

def _auto_bins(rsr, threshold):
    """自动计算分档区间"""
    rsr_sorted = np.sort(rsr.flatten())
    n = len(rsr_sorted)
    step = int(n * threshold)
    bins = [rsr_sorted[0] - 0.01]  # 左边界
    for i in range(step, n, step):
        bins.append(rsr_sorted[i-1])
    bins.append(rsr_sorted[-1] + 0.01)  # 右边界
    return np.unique(bins)

def _plot_rsr(result, bins):
    """绘制RSR分档可视化"""
    plt.figure(figsize=(10, 6))
    
    # RSR分布直方图
    plt.subplot(121)
    plt.hist(result['RSR值'], bins=20, edgecolor='k', alpha=0.7)
    for b in bins[1:-1]:
        plt.axvline(b, color='r', linestyle='--')
    plt.title('RSR Distribution with Bins')
    plt.xlabel('RSR Value')
    plt.ylabel('Frequency')
    
    # 分档结果条形图
    plt.subplot(122)
    grade_counts = result['分档'].value_counts().sort_index()
    grade_counts.plot(kind='bar', color='skyblue', edgecolor='black')
    plt.title('Grade Distribution')
    plt.xlabel('Grade')
    plt.ylabel('Count')
    
    plt.tight_layout()
    plt.show()

# 示例数据
data = np.array([
    [80, 70, 90, 85],  # 样本1
    [75, 85, 80, 78],  # 样本2
    [90, 60, 85, 92],  # 样本3
    [70, 90, 75, 80],  # 样本4
    [85, 75, 70, 88]   # 样本5
])

# 运行分析 (可自定义权重和方向)
result = rsr_analysis(
    data, 
    weight=[0.3, 0.2, 0.2, 0.3],  # 各指标权重
    ascending=[True, True, False, True]  # 第3个指标越小越好
)

print(result.sort_values('RSR值', ascending=False))
```
