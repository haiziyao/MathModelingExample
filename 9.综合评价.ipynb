型和评价者五个基本要素组成。**

**评价指标的筛选方法**
* 专家调研法
    * 当专家意见趋于集中时,将专家意见集中的指标作为评价指标,从而建立起综合评价指标体系
* 最小均方差法
    * 最小均方差法只考虑了指标的差异程度,容易将重要的指标删除
* 极大极小离差法
* 条件广义方差极小法
* 极大不相关法

**评价指标的预处理方法**

**一致性化**
* 极小型指标化为极大型指标
* 居中型指标化为极大型指标
* 区间型指标化为极大型指标

```python
import numpy as np
import matplotlib.pyplot as plt


plt.rcParams['font.sans-serif'] = ['SimHei']  # 使用黑体显示中文
plt.rcParams['axes.unicode_minus'] = False  # 解决负号显示问题

# 原始转换函数
def min2max(arr):
    """越小越好 -> 越大越好 (取补)"""
    arr = list(arr)
    maxx = max(arr)
    return np.array([[maxx - e] for e in arr])

def mid2max(mid, arr):
    """接近某值最好 -> 越大越好 (基于距离)"""
    arr = list(arr)
    gap = [abs(e - mid) for e in arr]
    M = max(gap) if max(gap) != 0 else 1
    return np.array([[1 - e / M] for e in gap])

def reg2max(left, right, arr):
    """区间最好 -> 越大越好 (区间外惩罚)"""
    arr = list(arr)
    M = max(left - min(arr), max(arr) - right) if max(left - min(arr), max(arr) - right) != 0 else 1
    ans = []
    for val in arr:
        if val < left:
            ans.append([1 - (left - val) / M])
        elif val > right:
            ans.append([1 - (val - right) / M])
        else:
            ans.append([1])
    return np.array(ans)

# 示例数据
original_data = np.array([3, 5, 8, 10, 12, 15, 18])

# 应用转换
data_min2max = min2max(original_data).flatten()  # 假设越小越好（如成本）
data_mid2max = mid2max(10, original_data).flatten()  # 假设10为最优值（如pH值）
data_reg2max = reg2max(8, 15, original_data).flatten()  # 假设8-15为最佳区间（如温度范围）

# 可视化对比
plt.figure(figsize=(12, 6))

plt.subplot(2, 1, 1)
plt.plot(original_data, 'o-', label='Original Data')
plt.title("Original Data vs Transformed Data")
plt.legend()

plt.subplot(2, 1, 2)
plt.plot(data_min2max, 's--', label='min2max (越小越好→越大越好)')
plt.plot(data_mid2max, 'd-.', label='mid2max (接近10最好)')
plt.plot(data_reg2max, '^:', label='reg2max (8-15区间最好)')
plt.axhline(y=1, color='gray', linestyle='--', linewidth=0.5)  # 理想值参考线
plt.legend()

plt.tight_layout()
plt.show()

# 打印转换结果
print("原始数据:\n", original_data)
print("\nmin2max转换（越小越好→越大越好）:\n", data_min2max)
print("\nmid2max转换（接近10最好）:\n", data_mid2max)
print("\nreg2max转换（8-15区间最好）:\n", data_reg2max)
```

**指标的无量纲化处理**

* 标准样本变换法
* 比例变换法
* 向量归一化
* 极差变换法
* 功效系数法


```python
import numpy as np
def standard_sample_transform(data):
    """
    标准样本变换法 (Z-score标准化)
    公式: (x - mean) / std
    """
    mean = np.mean(data, axis=0)
    std = np.std(data, axis=0)
    return (data - mean) / std
def ratio_transform(data):
    """
    比例变换法
    公式: x / max(x)
    """
    return data/np.max(np.abs(data))
def vector_normalization(data):
    """
    向量归一化法
    公式: x / sqrt(sum(x^2))
    """
    return data / np.sqrt(np.sum(data**2, axis=0))

def range_transform(data):
    """
    极差变换法 (Min-Max标准化)
    公式: (x - min) / (max - min)
    """
    min_val = np.min(data, axis=0)
    max_val = np.max(data, axis=0)
    return (data - min_val) / (max_val - min_val)

def efficacy_coefficient(data, a=60, b=40):
    """
    功效系数法
    公式: a + b * (x - min) / (max - min)
    默认参数: a=60, b=40 (将数据转换到60-100区间)
    """
    min_val = np.min(data, axis=0)
    max_val = np.max(data, axis=0)
    return a + b * (data - min_val) / (max_val - min_val)


# 示例数据
sample_data = np.array([
    [100, 0.5, 200],
    [80, 0.4, 180],
    [120, 0.6, 220],
    [90, 0.55, 210]
])
# 应用各种无量纲化方法
print("原始数据:\n", sample_data)
print("\n标准样本变换法:\n", standard_sample_transform(sample_data))
print("\n比例变换法:\n", ratio_transform(sample_data))
print("\n向量归一化法:\n", vector_normalization(sample_data))
print("\n极差变换法:\n", range_transform(sample_data))
print("\n功效系数法:\n", efficacy_coefficient(sample_data))
```

#### **常用的综合评价数学模型**
* **线性加权模型**:
  
    * 由于总的权重之和为1,各指标可以线性相互补偿;
    * 权重系数对评价结果的影响明显,权重大的指标对综合指标作用较大;
    * 计算简单,可操作性强;
    * 线性加权综合评价模型适用于各评价指标之间相互独立的情况,若m个评
    *价指标不完全独立,其结果将导致各指标间信息的重复起作用,使评价结果不能客观地反映实际*
* **TOPSOS法(理想解的排序方法)**

  它借助于评价问题的正理想解和负理想解,对各评价对象进行排序,所谓正理想解是一个虚拟的最佳对象,其每个指标值都是所有评价对象中该指标的最好值;而负理想解则是另一个虚拟的最差对象,其每个指标值都是所有评价对象中该指标的最差值. 求出各评价对象与正理想解和负理想解的距离,并以此对各评价对象进行优劣排序
    [查看典例](./Method_Reference/TOPSIS.ipynb)
* **灰色关联分析**
  
  根据灰色关联度的大小,对各评价对象进行排序,可建立评价对象的关联度,关联度越大其评价结果越好 [查看典例](./Method_Reference/GRE_GRA.ipynb)

* **熵值法**
  
    熵值法就是通过计算各指标观测值的信息 ,根据各指标的相对变化程度对系统整体的影响来确定指标权重系数的一种赋权方法[查看典例](./Method_Reference/EWM.ipynb)
* **秩和比法**

    秩和比(rank sum ratio, RSR) 综合评价法的基本原理是在一个n行m 列矩阵中,通过秩转换,获得无量纲统计量 RSR; 以RSR值对评价对象的优劣直接排序,从而对评价对象做出综合评价.[查看范例](./Method_Reference/RSR.ipynb)
