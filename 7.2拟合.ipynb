* **ployfit拟合**

```python
from numpy import polyfit,array,arange,polyval
from matplotlib.pyplot import plot,show,rc

rc('font',size=16)

x0=arange(0, 1.1, 0.1)
y0=array([-0.447, 1.978, 3.28, 6.16, 7.08, 7.34, 7.66, 9.56, 9.48,
9.30, 11.2])
p = polyfit(x0,y0,2) #2表示二次多项式进行拟合
print(f"系数：{p}")
x_topred = [0.25,0.35]
y_pred = polyval(p,x_topred)
print(f"预测结果{y_pred}")
plot(x0,y0,'*',x0,polyval(p,x0),'-')
```

* **curve_fit**


```python
import numpy as np
from scipy.optimize import curve_fit
y=lambda x, a, b, c: a*x**2+b*x+c
x0=np.arange(0, 1.1, 0.1)
y0=np.array([-0.447, 1.978, 3.28, 6.16, 7.08, 7.34, 7.66, 9.56,9.48, 9.30, 11.2])
popt, pcov=curve_fit(y, x0, y0)
print("拟合的参数值为:", popt)
print("预测值分别为:", y(np.array ([0.25,0.35]), *popt)) #这里的*是解包参数，将数组拆了
```
